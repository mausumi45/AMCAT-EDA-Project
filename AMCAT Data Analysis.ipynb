```python
# Define the search terms
specialization_terms = ['computer science & engineering', 'computer science and technology','computer science']
designation_terms = [ 'Programming Analyst', 'Software Engineer',  'Hardware Engineer', 'Associate Engineer']

# Step 1: Filter the DataFrame for Specialization
specialization_filter = Data['Specialization'].str.contains('|'.join(specialization_terms), case=False)

# Step 2: Filter the DataFrame for Designation
designation_filter = Data['Designation'].str.contains('|'.join(designation_terms), case=False)

# Step 3: Combine both filters
filtered_df = Data[specialization_filter & designation_filter]

# Display the filtered DataFrame
print("Filtered DataFrame:")
print(filtered_df)
# Step 5: Test Salary Claim
average_salary = filtered_df['Salary'].mean()
salary_claim_valid = 250000 <= average_salary <= 300000

# Display results for salary claim
print("Filtered DataFrame:")
print(filtered_df)
print(f"\nAverage Salary of the filtered designations: {average_salary}")
print(f"Does the average salary claim hold true? {'Yes' if salary_claim_valid else 'No'}")

```

```python
# Step 6: Analyze Gender and Specialization
gender_specialization_count = pd.crosstab(filtered_df['Gender'], filtered_df['Specialization'])
print("\nCrosstab of Gender and Specialization:")
print(gender_specialization_count)

# Calculate proportions
gender_specialization_proportions = gender_specialization_count.div(gender_specialization_count.sum(axis=1), axis=0)
print("\nProportions of Specialization by Gender:")
print(gender_specialization_proportions)

# Step 7: Visualization
plt.figure(figsize=(10, 6))
sns.barplot(data=gender_specialization_count.reset_index().melt(id_vars='Gender'), x='Gender', y='value', hue='Specialization')
plt.title('Count of Specialization by Gender', fontsize=16)
plt.xlabel('Gender', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.xticks(rotation=0)
plt.legend(title='Specialization')
plt.show()
```
